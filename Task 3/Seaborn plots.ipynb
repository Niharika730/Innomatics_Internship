```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel('D:/Class files/data/aspiring_minds_employability_outcomes_2015.xlsx')
df.head()
```

# Factor plot

# Observation:

*1.Factor plot is Bivariate.*

*2.Factor plot is both numerical and categorical.*

*3.In most of the observations 10percentage and 12percentage are almost correlated.*

```python
 with sns.axes_style(style='ticks'):
    g = sns.factorplot("10percentage", "12percentage", "Gender", data=df)
    g.set_axis_labels("10percentage", "12percentage");
```

# Violinplot

# Observation:


*1.Violinplot is Bivariate.*

*2.Violinplot is both numerical and Categorical.*

*3.Males are more compared to females.*

*4.For males 12percentage mostly lies between 60 t0 80.*

*5.For females 12percentage mostly lies between 70 t0 85.*

```python
sns.violinplot("Gender", "12percentage", data=df,
               palette=["lightblue", "lightpink"]);
```

# lmplot

# Observation:

*1.lmplot is Bivariate.*

*2.lmplot is numerical.*

*3.From the graph CollegeTier1 is very less compared collegetier1.*

```python
g = sns.lmplot('CollegeTier', '10percentage', col='Gender', data=df,
               markers=".", scatter_kws=dict(color='c'))
g.map(plt.axhline, y=0.1, color="k", ls=":");
```

# kde plot

# Observation:

*1.Kde plot is Univariate.*

*2.Kde plot is Numerical.*

*3.Salary hikes more between 0 to 1.*

```python
sns.kdeplot(df['Salary']);
```

# Stack plot

# Observation:


*1.Stack plot is Bivariate.*

*2.Stack plot is both Categorical and Numerical.*

*3.For both males and females 12percentage is mostly between 40 t0 100.*


```python
x=df['Gender']
y=df['12percentage']
pal = sns.color_palette("Set2")
plt.stackplot(x,y,colors=pal,labels=['A','B'], alpha=0.4 )
plt.legend(loc='upper right')
plt.show()
```

# Boxen plot:

# Observation:
*1.Boxen plot is bivariate.*


*2.Boxen plot contains both numeric and categorical data.*


*3.Most of the people are graduated in 2017.*

```python
sns.boxenplot(x = "GraduationYear", y = "CollegeTier",  
              data = df) 
plt.show()
```

# Cluster plot

```python
sns.clustermap(df.corr(),cmap='viridis')
```

# lmplot

# Observation:

*1.lmplot is Bivariate.*


*2.lmplot is numerical.*


*3.In Tier1 and Tier2 Salary lies between 0.5 to 1.5.*


*4.CollegeTier2 people are more compared to CollegeTier1.*

```python
sns.lmplot(x="Salary", y="CollegeTier",data=df, robust=True, ci=None, scatter_kws={"s": 80})
```

# Resid plot

# Observation:

*1.Residplot is Bivariate.*

*2.Residplot is numerical.*

*3.CollegeTier 2 people are more compared to CollegeTier1.*

```python
sns.residplot(x="CollegeTier", y="10percentage", data=df)
```

# relplot:

# Observation:


*1.Relplot is Bivariate.*



*2.Relplot is both categorical and numeric.*



*3.People from CollegeTier 2 are more than CollegeTier 1 irrespective of gender.*



*4.Males are more compared to females.*



*5.Minimum marks in 10percentage is between 45 to 50 and maximum marks is between 97 to 100.*
